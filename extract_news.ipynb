```python
import newspaper
#ndtv=newspaper.build('https://www.deccanherald.com/')
#for article in ndtv.articles:
#    print(article.url)

newso=newspaper.build('https://www.bbc.com/news/world/asia/india')
for article in newso.articles:
  print(article.url)
```

```python
from newspaper import Article

headlines = []
urls = []
contents = []
publication_dates = []
keywords = []
```

```python
for article in newso.articles:
    # Extract information from each article and append to respective lists
    art=Article(article.url, language="en")
    art.download()
    art.parse()
    urls.append(article.url)
    headlines.append(art.title)
    contents.append(art.text)
    publication_dates.append(art.publish_date)
    keywords.append(art.keywords)

# Create a DataFrame from the lists
df = pd.DataFrame({
    'Headline': headlines,
    'URL': urls,
    'Content': contents,
    'Publication_Date': publication_dates,
    'Keywords': keywords
})
```

```python
df.to_csv('News.csv')
```
